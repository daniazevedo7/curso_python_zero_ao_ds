## 1. Qual a média do preço de compra dos imóveis por “Nível”? 
        - Nível 0 -> Preço entre R$ 0 e R$ 321.950
        - Nível 1 -> Preço entre R$ 321.950 e R$ 450.000
        - Nível 2 -> Preço entre R$ 450.000 e R$ 645.000
        - Nível 3 -> Acima de R$ 645.000

```python
   #loop
for i in range(0, len (data)):
    if (data.loc[i, 'price'] > 0) & (data.loc[i, 'price'] < 321950):
        data.loc[i, 'nivel'] = 'nivel_0'

    elif (data.loc[i, 'price'] >= 321950) & (data.loc[i, 'price'] < 450000):
        data.loc[i, 'nivel'] = 'nivel_1'

    elif (data.loc[i, 'price'] >= 450000) & (data.loc[i, 'price'] < 645000):
        data.loc[i, 'nivel'] = 'nivel_2'
    else:
        data.loc[i, 'nivel'] = 'nivel_3'
```

```python
df = data[['price', 'nivel']].groupby('nivel').mean().reset_index()
df
```

## 4. Adicione os seguinte filtros no Mapa:
        - Tamanho mínimo da área da sala de estar.
        - Número mínimo de banheiros.
        - Valor Máximo do Preço.
        - Tamanho máximo da área do porão.
        - Filtro das Condições do Imóvel.
        - Filtro por Ano de Construção.

```python
data = pd.read_csv( '/home/daniel/repos/curso_python_zero_ao_ds/datasets/kc_house_data.csv' )
#data = pd.read_csv( 'C:/Users/daniel.azevedo/repos/curso_python_zero_ao_ds/datasets/kc_house_data.csv' )
```

```python
data.columns
```

```python
data[['sqft_living', 'bathrooms', 'price', 'sqft_basement', 'condition', 'yr_built']]
```

```python
#Interactive buttons

#Botão sala de estar
saladeestar_filtro = widgets.IntSlider(
    value=2079, 
    min=290,
    max = 13540,
    description = 'Área sala de estar',
    style={'description_width': 'initial'})

#Botão banheiro
banheiro_filtro = widgets.IntSlider(
    value=2,
    min=0,
    max=8,
    description = ('Número de banheiros'),
    orientation = 'vertical',
    style={'description_width': 'initial'})

#Botão preço
preco_filtro = widgets.IntSlider(
    value = data['price'].mean(),
    min = data['price'].min(),
    max = data['price'].max(),
    step = 1,
    description = 'Preços dos imóveis',
    disabled = False,
    orientation = 'horizontal',
    style={'description_width': 'initial'})

#Botão Porão
porao_filtro = widgets.IntSlider(
    value= data['sqft_basement'].mean(),
    min= data['sqft_basement'].min(),
    description = 'Área do porão',
    style={'description_width': 'initial'})

#Botão Condição imóvel
condicao_filtro = widgets.RadioButtons(
    options  = [1, 2 ,3, 4, 5],
    description = 'Condição do imóvel',
    disabled = False,
    style={'description_width': 'initial'})

#Botão ano de construção
ano_filtro = widgets.Dropdown(
    options = sorted(data['yr_built'].unique()),
    description = 'Ano de construção do imóvel',
    disabled = False,
    style={'description_width': 'initial'})

def update_map(df, saladeestar_filtro, banheiro_filtro, preco_filtro, porao_filtro,
               condicao_filtro, ano_filtro):
    houses = df[(df['sqft_living'] > saladeestar_filtro) &
                (df['bathrooms'] > banheiro_filtro) &
                (df['price']< preco_filtro) &
                (df['sqft_basement'] > porao_filtro) &
                (df['condition'] > condicao_filtro) &
                (df['yr_built'] > ano_filtro)][['id', 'lat', 'long', 'price', 'nivel']]
    
    fig = px.scatter_mapbox(houses,
                           lat='lat',
                           lon='long',
                           size='price',
                           color_continuous_scale=px.colors.cyclical.IceFire,
                           zoom=10)
    
    fig.update_layout(mapbox_style="open-street-map")
    fig.update_layout(height=600, margin={"r":0,"t":0,"l":0,"b":0})
    fig.show()
    
```

```python
widgets.interactive(update_map,
                    df=fixed(data),
                    ano_filtro = ano_filtro,
                    condicao_filtro = condicao_filtro,
                    porao_filtro = porao_filtro,
                    banheiro_filtro = banheiro_filtro,
                    saladeestar_filtro = saladeestar_filtro,
                    preco_filtro=preco_filtro,
                   )
```
